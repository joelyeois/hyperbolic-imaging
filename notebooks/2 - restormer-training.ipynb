```python
import torch
import matplotlib.pyplot as plt
import sys
sys.path.append('../src/')

from restormer import DataModule, Restormer
import pytorch_lightning as pl

%load_ext autoreload
%autoreload 2
```

```python
torch.set_default_dtype(torch.float32)
```

## Load data

```python
x_clean = torch.load('../data/training_dataset/groundtruth_images.pt')
x_blur = torch.load('../data/training_dataset/blurred_images.pt')
x_clean.shape, x_blur.shape
```

```python
#expand channels
x_clean = x_clean[:,None,:,:]
x_blur = x_blur[:,None,:,:]
```

```python
#view some images from dataset
fig, axes = plt.subplots(2,5, dpi=200, figsize=(8,3.15), constrained_layout=True)
[ax.set(xticks=[], yticks=[]) for ax in axes.ravel()]

for i in range(5):
    ax = axes[0,i]
    ax.imshow(torch.squeeze(x_clean[i][0]), cmap='gray')
    ax = axes[1,i]
    ax.imshow(torch.squeeze(x_blur[i][0]), cmap='gray')
axes[0,0].set_ylabel('clean')
axes[1,0].set_ylabel('blur & noisy')
plt.show()
```

## Restormer deblurring
PyTorch Lightning Trainer automatically saves model checkpoints in './lightning_logs/version_.../checkpoints/epoch=...'

```python
#initiate model
niter = 200
chn = 36 #reduce this if not enough GPU memory
batch_size = 1
model = Restormer(n_channels=1, niter=niter, channels=[chn, chn*2, chn*4, chn*8])
```

```python
N = 3500 #use 87.5% of data as training
dm = DataModule(batch_size, x_blur[:N], x_clean[:N])
dm.prepare_data()
```

```python
#Change accelerator and devices to suit your workstation specs.
trainer = pl.Trainer(accelerator='gpu', 
                     devices=[0,1,2,3], #gpu numbers.
                     max_epochs=niter)
trainer.fit(model, dm)
```
